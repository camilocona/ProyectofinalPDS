```python
import matplotlib.pyplot as plt

# Definir el valor del polo
s_polo = -1 / (1000 * 15.9 * 10**-6)

# Extraer la parte real e imaginaria del polo
real_part = s_polo.real
imaginary_part = s_polo.imag

# Graficar el diagrama de polos y ceros
plt.figure(figsize=(6, 6))
plt.plot(real_part, imaginary_part, 'rx', markersize=10)  # Polo
plt.axvline(x=0, color='k', linestyle='--')  # Eje imaginario
plt.axhline(y=0, color='k', linestyle='--')  # Eje real
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginaria')
plt.title('Diagrama de Polos y Ceros')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Definir la función de transferencia H(s)
def transfer_function(s, R, C):
    return 1 / (1 + s * R * C)


# Parámetros del circuito
R = 1000  # Resistencia en ohmios
C = 15.9 * 10**-6  # Faradios

# Crear un rango de valores para s en el plano complejo
s_values = np.linspace(-10000, 10000, 1000) * 1j  # Valores de s

# Calcular la respuesta en frecuencia de H(s)
H_values = transfer_function(s_values, R, C)

# Graficar la magnitud y la fase de H(s)
plt.figure(figsize=(10, 5))

# Magnitud de H(s)
plt.subplot(2, 1, 1)
plt.plot(s_values.imag, np.abs(H_values))
plt.title('Respuesta en frecuencia de la función de transferencia H(s)')
plt.xlabel('Parte Imaginaria de s')
plt.ylabel('Magnitud')
plt.grid(True)


plt.tight_layout()
plt.show()
```
